```python
import pandas as pd
import plotly.express as px


# Dataset: Gapminder — GDP, Life Expectancy, Population by Country
# Source: Gapminder Foundation (gapminder.org)

df = px.data.gapminder()
print(f"Loaded: {len(df)} rows")
print(df.head())
```

## Task 1 — Scatter: life expectancy change over time

**What to build:** A scatter showing **GDP per capita vs life expectancy** for **two years** (2000 and 2007) to show how both moved — use **colour for year** (just 2 colours), **one continent only**.

Choose any continent except Africa (that was the example). Highlight the change direction.

> 💡 Filter: `df.loc[df['continent'] == 'YOUR_CHOICE']` then filter years


```python
import plotly.graph_objects as go
import plotly.io as pio

# Use browser renderer to avoid nbformat issues
pio.renderers.default = "browser"

# ---------------------------------------------------------
# Filter to one continent (Asia) and two years
# ---------------------------------------------------------
asia = df.loc[(df['continent'] == 'Asia') & (df['year'].isin([2000, 2007]))]

# Split by year
asia_2000 = asia[asia['year'] == 2000]
asia_2007 = asia[asia['year'] == 2007]

fig = go.Figure()

# ---------------------------------------------------------
# Scatter for 2000
# ---------------------------------------------------------
fig.add_trace(go.Scatter(
    x=asia_2000['gdpPercap'],
    y=asia_2000['lifeExp'],
    mode='markers',
    marker=dict(size=10, color='#1E88E5'),
    name='2000'
))

# ---------------------------------------------------------
# Scatter for 2007
# ---------------------------------------------------------
fig.add_trace(go.Scatter(
    x=asia_2007['gdpPercap'],
    y=asia_2007['lifeExp'],
    mode='markers',
    marker=dict(size=10, color='#D81B60'),
    name='2007'
))

# ---------------------------------------------------------
# Add arrows showing direction of change
# ---------------------------------------------------------
for country in asia['country'].unique():
    p2000 = asia_2000[asia_2000['country'] == country]
    p2007 = asia_2007[asia_2007['country'] == country]

    if not p2000.empty and not p2007.empty:
        fig.add_annotation(
            x=p2007['gdpPercap'].values[0],
            y=p2007['lifeExp'].values[0],
            ax=p2000['gdpPercap'].values[0],
            ay=p2000['lifeExp'].values[0],
            xref='x', yref='y', axref='x', ayref='y',
            showarrow=True,
            arrowhead=3,
            arrowsize=1,
            arrowwidth=1.5,
            arrowcolor='gray'
        )

# ---------------------------------------------------------
# Layout
# ---------------------------------------------------------
fig.update_layout(
    title="Asia’s Progress: Rising GDP and Longer Lives (2000 → 2007)",
    xaxis_title="GDP per Capita",
    yaxis_title="Life Expectancy",
    plot_bgcolor="white",
    height=550
)

fig.show()


```

## Task 2 — Bubble chart: tell a story

**What to build:** A bubble chart (full 2007 dataset, all countries) where:
- x = GDP per capita (log scale)
- y = life expectancy
- size = population
- colour = ONE continent highlighted (your choice), all others grey
- At least one annotation explaining the highlighted group's story

> This is the grey-and-highlight technique applied to a bubble chart.


```python
# Task 2
# YOUR CODE HERE

# Task 2 — Bubble chart: tell a story

import plotly.express as px

# Filter for the year 2007
df_2007 = df[df['year'] == 2007]

# Choose one continent to highlight (e.g., Asia)
highlight = 'Asia'

# Create a new column for colour: highlight one continent, grey out others
df_2007['highlight'] = df_2007['continent'].apply(
    lambda x: highlight if x == highlight else 'Other'
)

# Plot the bubble chart
fig = px.scatter(
    df_2007,
    x='gdpPercap',
    y='lifeExp',
    size='pop',
    color='highlight',
    color_discrete_map={highlight: 'crimson', 'Other': 'lightgrey'},
    hover_name='country',
    size_max=60,
    log_x=True,
    title=f"Life Expectancy vs GDP per Capita (2007) — Highlighting {highlight}"
)

# Add annotation explaining the highlighted group's story
fig.add_annotation(
    x=12000,
    y=75,
    text=f"{highlight} shows rapid economic growth and rising life expectancy.",
    showarrow=True,
    arrowhead=2
)

fig.show()
```
